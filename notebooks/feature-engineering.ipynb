```python
import pandas as pd
import numpy as np

df_fraud=pd.read_csv("../data/processed/fraud_final.csv")
df_fraud.sample(3)
```

```python
# =====================================
# Frequency Encoding for device_id
# =====================================

# 1. Check cardinality (optional but good practice)
print("Unique device_ids:", df_fraud['device_id'].nunique())

# 2. Compute frequency of each device_id
device_freq = (
    df_fraud['device_id']
    .value_counts(normalize=True)
)

# 3. Map frequencies to a new column
df_fraud['device_id_freq'] = (
    df_fraud['device_id']
    .map(device_freq)
)

# 4. Sanity checks
print("Missing after encoding:", df_fraud['device_id_freq'].isna().sum())
print(df_fraud[['device_id', 'device_id_freq']].head())

# 5. Drop original device_id
df_fraud.drop(columns=['device_id'], inplace=True)
```

```python
df_fraud = df_fraud.to_csv("../data/processed/fraud_final.csv", index=False)
print ("Final processed data saved to fraud_final.csv")
```

```python
# =====================================
# Frequency Encoding for country
# =====================================

if 'country' in df_fraud.columns:
    # 1. Check number of unique countries
    print("Unique countries:", df_fraud['country'].nunique())

    # 2. Compute frequency distribution
    country_freq = (
        df_fraud['country']
        .value_counts(normalize=True)
    )

    # 3. Map frequencies to new column
    df_fraud['country_freq'] = (
        df_fraud['country']
        .map(country_freq)
    )

    # 4. Sanity checks
    print("Missing after encoding:", df_fraud['country_freq'].isna().sum())
    print(df_fraud[['country', 'country_freq']].head())

    # 5. Drop original country column
    df_fraud.drop(columns=['country'], inplace=True)
else:
    print("Country column already processed or not present.")
```
